#                Matplotlib

### Jenis-jenis Plot


##### 1. Plot Garis 


```python
from matplotlib import pyplot as plt

x = [4,6,10,13,16,20,22,24,26,30]
y = [100,120,160,180,210,270,300,320,330,350]
plt.plot(x,y)
plt.show()
```

##### Label dan nama

```python
# Label dan Title

x = [4,6,10,13,16,20,22,24,26,30]
y = [100,120,160,180,210,270,300,320,330,350]
plt.plot(x,y)

# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")
plt.show()
```

##### Multiline

```python
# Multiline
hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.plot(hari, solehudin)

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.plot(hari, sientong)


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")
plt.show()
```

##### Legend

```python
# Legend
hari= [4,6,10,13,16,20,22,24,26,30]

#sumbu y pertama
solehudin = [100,120,160,180,210,270,300,320,330,350]
plt.plot(hari, solehudin)

#sumbu y kedua
sientong = [70,80,110,130,160,180,200,220,280,300]
plt.plot(hari, sientong)


# Menambah Label dan Title
plt.title("Kenaikan Subcriber berdasarkan hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah Subcriber")

# Legend
plt.legend(['solehudin','sientong'])
plt.show()
```
